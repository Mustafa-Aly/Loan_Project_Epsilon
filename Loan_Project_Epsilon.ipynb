###**Modeling**

```python
#pip install --upgrade category_encoders
```

```python
from category_encoders import BinaryEncoder
from sklearn.preprocessing import MinMaxScaler,StandardScaler,RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score , recall_score , precision_score , confusion_matrix , ConfusionMatrixDisplay
```

```python
df['revol.bal']=df['revol.bal'].apply(lambda x:x if x==0 else np.log(x))
```

```python
df['revol.bal'].value_counts()
```

```python
be=BinaryEncoder()
df_be=be.fit_transform(df['purpose'])
```

```python
df=pd.concat([df,df_be],axis=1)
```

```python
df.drop(['purpose'],axis=1,inplace=True)
```

```python
x,y=df.drop(['not.fully.paid'],axis=1),df['not.fully.paid']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5,stratify=y)
```

```python
y_train.value_counts()
```

```python
smt = SMOTE(sampling_strategy={1 : 3000})
```

```python
x_train , y_train = smt.fit_resample(x_train , y_train)
```

```python
sclr=RobustScaler()
#sclr=MinMaxScaler()
```

```python
x_train=sclr.fit_transform(x_train)
x_test=sclr.transform(x_test)
```

```python
x_train
```

```python
models=[
    ('lr',LogisticRegression()),
    ('SVM',SVC()),
    ('NB',GaussianNB()),
    ('KNN',KNeighborsClassifier()),
    ('DT',DecisionTreeClassifier()),
    ('RF',RandomForestClassifier()),
    ('XG',XGBClassifier())
]
```

```python
for Model in models:
  Model[1].fit(x_train,y_train)


  y_train_pred=Model[1].predict(x_train)
  y_test_pred=Model[1].predict(x_test)


  print(Model[0])

  print(f'Training score:{accuracy_score(y_true=y_train,y_pred=y_train_pred)}')
  print(f'Testing score:{accuracy_score(y_true=y_test,y_pred=y_test_pred)}')

  print(f'Train recall : {recall_score(y_true=y_train , y_pred=y_train_pred)}')
  print(f'Test recall : {recall_score(y_true=y_test , y_pred=y_test_pred)}')

  print(f'Train precision : {precision_score(y_true=y_train , y_pred=y_train_pred)}')
  print(f'Test precision : {precision_score(y_true=y_test , y_pred=y_test_pred)}')

  print('*'*50)
```

Decision Tree

```python
dtc = DecisionTreeClassifier(max_depth=12,max_features=5)
```

```python
dtc.fit(x_train , y_train)
```

```python
y_train_prd = dtc.predict(x_train)
y_test_prd = dtc.predict(x_test)

print(f'Train Acc : {accuracy_score(y_true=y_train , y_pred=y_train_prd)}')
print(f'Test Acc : {accuracy_score(y_true=y_test , y_pred=y_test_prd)}')

print(f'Train recall : {recall_score(y_true=y_train , y_pred=y_train_prd)}')
print(f'Test recall : {recall_score(y_true=y_test , y_pred=y_test_prd)}')

print(f'Train precision : {precision_score(y_true=y_train , y_pred=y_train_prd)}')
print(f'Test precision : {precision_score(y_true=y_test , y_pred=y_test_prd)}')
```
